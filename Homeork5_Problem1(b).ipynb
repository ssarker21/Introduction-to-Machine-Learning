```python
import torch
import matplotlib.pyplot as plt

t_c = [0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c, dtype=torch.float32)
t_u = torch.tensor(t_u, dtype=torch.float32)

# Scale t_u (your code)
t_u = t_u * 0.15

def model_quadratic(t_u, w2, w1, b):
    return w2 * t_u**2 + w1 * t_u + b

def loss_fn(t_p, t_c):
    squared_diffs = (t_p - t_c)**2
    return squared_diffs.mean()

def training_loop(n_epochs, optimizer, params, t_u, t_c):
  history = []                                   # <-- minimal addition
  for epoch in range(1, n_epochs + 1):
    t_p = model_quadratic(t_u, *params)
    loss = loss_fn(t_p, t_c)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    history.append(float(loss.detach()))         # <-- minimal addition
    if epoch % 500 == 0:
      print('Epoch %d, loss %f' % (epoch, float(loss)))
  return params, history                         # <-- minimal addition

# lr = 1e-1
params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
optimizer = torch.optim.SGD([params], lr=1e-1)
_, hist_1e1 = training_loop(5000, optimizer, params, t_u, t_c)

# lr = 1e-2
params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
optimizer = torch.optim.SGD([params], lr=1e-2)
_, hist_1e2 = training_loop(5000, optimizer, params, t_u, t_c)

# lr = 1e-3
params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
optimizer = torch.optim.SGD([params], lr=1e-3)
_, hist_1e3 = training_loop(5000, optimizer, params, t_u, t_c)

# lr = 1e-4
params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
optimizer = torch.optim.SGD([params], lr=1e-4)
_, hist_1e4 = training_loop(5000, optimizer, params, t_u, t_c)

# ---- FOUR separate HD and BOLD plots (one per learning rate) ----
plt.figure(figsize=(7, 4.2), dpi=300)
plt.plot(hist_1e1, linewidth=3, color='blue')
plt.xlabel('Epoch', fontsize=12, fontweight='bold')
plt.ylabel('Loss (MSE)', fontsize=12, fontweight='bold')
plt.title('SGD: lr = 1e-1', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4.2), dpi=300)
plt.plot(hist_1e2, linewidth=3, color='green')
plt.xlabel('Epoch', fontsize=12, fontweight='bold')
plt.ylabel('Loss (MSE)', fontsize=12, fontweight='bold')
plt.title('SGD: lr = 1e-2', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4.2), dpi=300)
plt.plot(hist_1e3, linewidth=3, color='red')
plt.xlabel('Epoch', fontsize=12, fontweight='bold')
plt.ylabel('Loss (MSE)', fontsize=12, fontweight='bold')
plt.title('SGD: lr = 1e-3', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4.2), dpi=300)
plt.plot(hist_1e4, linewidth=3, color='purple')
plt.xlabel('Epoch', fontsize=12, fontweight='bold')
plt.ylabel('Loss (MSE)', fontsize=12, fontweight='bold')
plt.title('SGD: lr = 1e-4', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


# ---- 4 separate plots (one per learning rate) ----
#plt.figure(); plt.plot(hist_1e1); plt.xlabel('Epoch'); plt.ylabel('Loss (MSE)'); plt.title('lr = 1e-1'); plt.tight_layout(); plt.show()
#plt.figure(); plt.plot(hist_1e2); plt.xlabel('Epoch'); plt.ylabel('Loss (MSE)'); plt.title('lr = 1e-2'); plt.tight_layout(); plt.show()
#plt.figure(); plt.plot(hist_1e3); plt.xlabel('Epoch'); plt.ylabel('Loss (MSE)'); plt.title('lr = 1e-3'); plt.tight_layout(); plt.show()
#plt.figure(); plt.plot(hist_1e4); plt.xlabel('Epoch'); plt.ylabel('Loss (MSE)'); plt.title('lr = 1e-4'); plt.tight_layout(); plt.show()
```
